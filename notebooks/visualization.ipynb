```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Ice Cream Ratings.csv")
```

```python
# This code sets the 'Date' column as the index of the DataFrame.
df = df.set_index('Date')
df
```

```python
#Plottng line graph
df.plot(kind='line',figsize=(10, 4), 
        title='Ice Cream Ratings Over Time',
        ylabel='Ratings', xlabel='Date')
```

```python
# Plottng bar graph
df.plot(kind='barh',figsize=(8, 4), 
        title='Ice Cream Ratings Over Time',
        ylabel='Ratings', xlabel='Date',stacked=True)
```

```python
# Plottng histogram
df.plot(kind='hist',figsize=(8, 4), 
        title='Ice Cream Ratings Over Time',
        ylabel='Ratings', xlabel='Date', bins=10, edgecolor='black')
```

```python
df.plot(kind='box',figsize=(8, 4), 
        title='Ice Cream Ratings Over Time')
```

```python
# Plottng area graph
df.plot(kind='area',figsize=(8, 4), 
        title='Ice Cream Ratings Over Time',
        ylabel='Ratings', xlabel='Date',stacked=True)
```

```python
# Plot a pie chart for the 'Flavor Rating' column
df['Flavor Rating'].plot(kind='pie', figsize=(8, 6), 
        title='Flavor Rating Distribution', autopct='%1.1f%%', legend=True, ylabel='')
```
