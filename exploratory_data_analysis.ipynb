```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('world_population.csv')
df
```

```python
df.describe()
```

```python
#This code sums the number of null values in each column of the DataFrame.
df.isnull().sum()
```

```python
#This code finds the number of unique values in each column of the DataFrame.
df.nunique()
```

```python
#This code sorts the DataFrame by the "World Population Percentage" column in descending order 
#and returns the top 10 rows.
df.sort_values(by="World Population Percentage", ascending=False).head(10)
```

```python
# Select only numeric columns for correlation
df.select_dtypes(include='number').corr()
```

```python
# Select only numeric columns for correlation and plot a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Numeric Columns')
```

```python
# This code groups the DataFrame by 'Continent' 
# and calculates the mean of numeric columns, sorting by '2022 Population' in descending order.
df.groupby('Continent')[df.select_dtypes(include='number').columns].mean().sort_values(by="2022 Population", ascending=False)
```

```python
#This code filters the DataFrame to show only rows where the 'Continent' column contains 'Oceania'.
df[df['Continent'].str.contains('Oceania')]
```

```python
#This code groups the DataFrame by 'Continent' and calculates the mean of population columns,
#sorting by '2022 Population' in descending order.
df2 = df.groupby('Continent')[['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population']].mean().sort_values(by="2022 Population",ascending=False)
df2
```

```python
#This code transposes the DataFrame df2, switching rows and columns.
df3 = df2.transpose()
df3
```

```python
# Plotting the transposed DataFrame
plt.figure(figsize=(12, 6))
df3.plot()
```
