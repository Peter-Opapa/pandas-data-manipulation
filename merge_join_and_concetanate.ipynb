**Merge and Concetanate**

```python
import pandas as pd
```

```python
df=pd.read_csv('LOTR.csv')
df
```

```python
df1=pd.read_csv('LOTR 2.csv')
df1
```

```python
#Merge df1 with df on 'FellowsipID' column
df_merged = pd.merge(df, df1, on=['FellowshipID','FirstName'], how='inner')
df_merged
```

```python
#This code performs outer join on 'FellowshipID' and 'FirstName' columns
df_outer = pd.merge(df, df1, on=['FellowshipID','FirstName'], how='outer')
df_outer
```

```python
#This code performs a left join on 'FellowshipID' and 'FirstName' columns
df_left = pd.merge(df, df1, on=['FellowshipID','FirstName'], how='left')
df_left
```

```python
#This code performs a right join on 'FellowshipID' and 'FirstName' columns
df_right = pd.merge(df, df1, on=['FellowshipID','FirstName'], how='right')
df_right
```

```python
#This code performs a cross join between df and df1
df_cross = pd.merge(df, df1, how='cross')
df_cross
```

```python
#Appending df1 to df
df_appended = pd.concat([df, df1], ignore_index=True)
df_appended
```
